# Reading csv file

```python
import pandas as pd

file = 'C:/Users/amits/Downloads/Mitre_Website_Scraped.csv'
df = pd.read_csv(file)
df
```

```python
X = df["Description"]
Y = df["ID"]
```

```python
ZZ = set(Y)
ZZ = list(ZZ)
Num_Classes = len(ZZ)
```

```python
X
```

# Text preprocessing

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
import re
import nltk
nltk.download('stopwords')    # only keeps alphabets
```

```python
X_New = []

for i in X:
  review = re.sub(r'[^a-zA-Z0-9\s.,!?:/()\[\]@_-]+', '', i)
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)

  X_New.append(review)
```

```python
X_New
```

# Classifier

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(Y)
```

```python
import pickle

with open('LabelEncoder.pkl', 'wb') as file:
    pickle.dump(le, file)
```

```python
from sklearn.metrics import accuracy_score
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Bidirectional, LSTM, Dropout, Embedding, GlobalMaxPooling1D
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
```

# Train - Test Split

```python
X_train_Word, X_test_Word, Y_train_Word, Y_test_Word = train_test_split(X_New, Y, test_size=0.3, random_state=42)
```

```python
Y_train_Word_Encoded = le.transform(Y_train_Word)
Y_test_Word_Encoded = le.transform(Y_test_Word)
```

```python
#Y_train_Encoded = le.transform(Y_train_Word)
#Y_test_Encoded = le.transform(Y_test_Word)
```

```python
Y_Word_Encoded = le.transform(Y)
```

# Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout

# Assuming max_words is the maximum number of words in your vocabulary
max_words = 10000

# Create the neural network model
model = Sequential()

# Add an Embedding layer
model.add(Embedding(input_dim=max_words, output_dim=100))

# Add a Bidirectional LSTM layer
model.add(Bidirectional(LSTM(64, return_sequences=True)))

# Add another Bidirectional LSTM layer
model.add(Bidirectional(LSTM(32)))

# Add Dropout for regularization
model.add(Dropout(0.5))

# Add a Dense layer with softmax activation for classification
model.add(Dense(Num_Classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Assuming you have already done label encoding for Y_train_Word_Encoded

# Convert text data to sequences
tokenizer = Tokenizer(num_words=max_words)
#tokenizer.fit_on_texts(X_train_Word)
```

# Training

```python
#sequences = tokenizer.texts_to_sequences(X_train_Word)
tokenizer.fit_on_texts(X_New)
sequences = tokenizer.texts_to_sequences(X_New)


# Padding sequences
X_train_padded = pad_sequences(sequences)

# Now, X_train_padded is a NumPy array, and Y_train_Word_Encoded is encoded labels


model.fit(X_train_padded, Y_Word_Encoded, epochs=10, batch_size=32)
```

```python
with open('Tokenizer.pkl', 'wb') as file:
    pickle.dump(tokenizer, file)
```

# Prediction

```python
sequences = tokenizer.texts_to_sequences(X_test_Word)

# Padding sequences
X_test_padded = pad_sequences(sequences)


LSTMPred = model.predict(X_test_padded)
```

```python
LSTMpredicted_classes = np.argmax(LSTMPred, axis=1)
```

```python
Y_test_Word_Encoded
```

```python
LSTMpredicted_classes
```

```python
accuracyLSTM = accuracy_score(Y_test_Word_Encoded, LSTMpredicted_classes)
print(accuracyLSTM)
```
